```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Plot 1

how the latency depends on N (ranging), for each alpha, with tle fixed in tle_min=5

N: 3, 10, 25, 50, 75, 100

alpha: 0, 0.1, 1.0

```python
data1 = pd.read_csv('../data_plot1.csv', sep=';')
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='n', y='avg_latency', hue='alpha', data=data1, marker='o', palette={0: 'blue', 0.1: 'green', 1: 'red'})
plt.xlabel('N')
plt.ylabel('Average Latency (ms)')
plt.title('N vs Latency for different alpha values and tle=5ms')
plt.legend(title='alpha')
plt.grid()
plt.show()
```
